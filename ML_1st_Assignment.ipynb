```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dic={
    'studyhour':[0,1,2,3,4,5,6,7],
    'gpa':[0,0.7,1.0,2.3,2.4,2.7,3.3,3.6]
}
df=pd.DataFrame(dic)


x=np.array(df[['studyhour']])
y=np.array(df[['gpa']])
x,y

plt.xlabel('studyhour')
plt.ylabel('gpa')
plt.scatter(x,y)
x_mean=np.mean(x)
y_mean=np.mean(y)
print("X mean is ",x_mean,"||| Y mean is ",y_mean)

num=0
den=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print("Intercept is ",float(b0),"Coefficient is ",float(b1))

y_pred=b0+b1*x
y_pred

plt.scatter(x,y)
plt.plot(x,y_pred,c="blue")
plt.show

# R-square=1-(SSR/SST)
num_r=0
den_r=0
for i in range(len(x)):
    num_r+=(y[i]-y_pred[i])**2
    den_r+=(y[i]-y_mean)**2
r_sq=1-(num_r/den_r)
r_sq
```
